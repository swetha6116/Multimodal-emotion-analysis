```python
import os
import random
import matplotlib.pyplot as plt
from keras.preprocessing import image

def load_images_from_path(path, label):
    images = []
    labels = []

    # List all files in the directory
    files = os.listdir(path)
    # Randomly select 8 files if there are more than 8
    selected_files = random.sample(files, min(8, len(files)))

    for file in selected_files:
        images.append(image.img_to_array(image.load_img(os.path.join(path, file), target_size=(224, 224, 3))))
        labels.append(label)

    return images, labels

def show_images(images, label):
    fig, axes = plt.subplots(1, len(images), figsize=(20, 3), subplot_kw={'xticks': [], 'yticks': []})
    fig.suptitle(label, fontsize=16)
    for i, ax in enumerate(axes.flat):
        ax.imshow(images[i] / 255)
    plt.show()

# Assuming emotions_classes is defined elsewhere
for key, value in emotions_classes.items():
    images, labels = load_images_from_path(os.path.join(AUDIO_SPECTROGRAMS_DIR, value), key)
    show_images(images, value)

```
